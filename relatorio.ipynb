## Descrição do Projeto

Nosso projeto consiste em facilitar o acesso aos dados do site https://www.publishnews.com.br/ranking

Como motivação, buscamos entre os integrantes do grupo pontos de interesse, encontramos em comum o hábito de ler.

O site Publish News apresenta relatórios semanais, mensais e anuais sobre os livros mais vendidos através de uma amostra enviada por diversos lojistas/livrarias. 
O ranking dos livros mais vendidos está disponível para visualização ao público mas não possui arquivos para download, impossibilitando um fácil para análises.

A partir disso, criamos uma classe extratora que fará todo esse trabalho, extraindo as informações, gerando tabelas e por fim salvando os arquivos em excel.

### Criação da Classe Extratora Publish News, Publicação no Pypi e Repertório de Código Git

Primeiro criamos uma classe que será usada para inicializar o webdriver, escolhemos um método que funcionará para qualquer pessoa, mesmo que esta não possua o webdriver instalado ou atualizado em seu computador. Esta classe recebe o nome IniciarDriver

Uma vez que selenium e webdriver estão funcionando, partimos para a parte extratora, foi necessário examinar o site e criar um plano para que os passos necessários antes de chegar nos dados desejados acontecessem de forma automatizada. Temos então a classe ExtracaoPublishNews, para ela funcionar utilizamos  a classe IniciarDriver como relação de herança para iniciar o construtor.

Aqui temos um breve resumo sobre os métodos que criamos e o que eles fazem. 
O método seleciona_anual não recebe nenhum parâmetro, ele apenas clica na aba para selecionarmos o ano dentro do site;
seleciona_xpath, criamos um dicionário onde o ano é a chave e o x path é o valor, foi necessário para identificar o elemento x path para o ano escolhido. Também fazemos o tratamento de exceções para o caso de o ano escolhido não existir; 
temos o método coleta_dados, onde extraímos todas as informações disponíveis para cada livro, já aproveitando para fazer os tratamentos dos resultados, guardamos em uma lista e criamos o dataframe;
o método salvar_dataframe, para salvar em arquivo xlsx;
e por fim, fecha_browser, para fechar o chrome.

Utilizamos o cookiecutter para facilitar na documentação do projeto, e melhorar a experiência de outros usuários. Tox e Travis para verificarmos e testarmos o projeto antes de subi-lo no Pypi e no Github.
